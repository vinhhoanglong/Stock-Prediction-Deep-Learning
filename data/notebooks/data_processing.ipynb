```python
from helper import *
import pandas as pd
import numpy as np
```

```python
data = pd.DataFrame()
```

```python
df_APPLE = pd.read_csv("../raw/APPLE.csv")
df_IBM = pd.read_csv("../raw/IBM.csv")
df_INTC = pd.read_csv("../raw/INTC.csv")
df_MSFT = pd.read_csv("../raw/MSFT.csv")
df_QCOM = pd.read_csv("../raw/QCOM.csv")
df_SPY = pd.read_csv("../raw/SPY.csv")
df_TXN = pd.read_csv("../raw/TXN.csv")
```

```python
df_dict = {
    'APPLE': df_APPLE,
    'IBM': df_IBM,
    'INTC': df_INTC,
    'MSFT': df_MSFT,
    'QCOM': df_QCOM,
    'SPY': df_SPY,
    'TXN': df_TXN
}
```

```python
for ticker, df in df_dict.items():
    df[f'pct_change_{ticker}'] = (df['Close'] - df['Close'].shift(1)) / df['Close'].shift(1)
    df[f'log_change_{ticker}'] = np.log(df['Close'] / df['Close'].shift(1))
```

```python
for ticker, df in df_dict.items():
    df[f'7ma_{ticker}'] = EMA(df['Close'], 7)
    df[f'14ma_{ticker}'] = EMA(df['Close'], 14)
    df[f'21ma_{ticker}'] = EMA(df['Close'], 21)
    
    df[f'7macd_{ticker}'] = MACD(df['Close'], 3, 11, 7)
    df[f'14macd_{ticker}'] = MACD(df['Close'], 7, 21, 14)
    
    df[f'7rsi_{ticker}'] = RSI(df['Close'], 7)
    df[f'14rsi_{ticker}'] = RSI(df['Close'], 14)
    df[f'21rsi_{ticker}'] = RSI(df['Close'], 21)
    
    df[f'7atr_{ticker}'] = atr(df['High'], df['Low'], 7)
    df[f'14atr_{ticker}'] = atr(df['High'], df['Low'], 14)
    df[f'21atr_{ticker}'] = atr(df['High'], df['Low'], 21)
    
    df[f'7upper_{ticker}'], df[f'7lower_{ticker}'] = bollinger_band(df['Close'], 7)
    df[f'14upper_{ticker}'], df[f'14lower_{ticker}'] = bollinger_band(df['Close'], 14)
    df[f'21upper_{ticker}'], df[f'21lower_{ticker}'] = bollinger_band(df['Close'], 21)
    
    df[f'7rsv_{ticker}'] = rsv(df['Close'], 7)
    df[f'14rsv_{ticker}'] = rsv(df['Close'], 14)
    df[f'21rsv_{ticker}'] = rsv(df['Close'], 21)
```

```python
for ticker, df in df_dict.items():
    fft_result = np.fft.fft(df['Close'])
    

    df[f'fft_real_{ticker}'] = fft_result.real
    df[f'fft_imag_{ticker}'] = fft_result.imag
    

    df[f'fft_magnitude_{ticker}'] = np.abs(fft_result)
    df[f'fft_phase_{ticker}'] = np.angle(fft_result)
```

```python
for ticker, df in df_dict.items():
    df['Date'] = pd.to_datetime(df['Date'])
    df.set_index('Date', inplace=True)

df_train_dict = {}


common_dates = pd.Index(df_dict['APPLE'].index)  
for ticker, df in df_dict.items():
    common_dates = common_dates.intersection(df.index)


for ticker, df in df_dict.items():

    df_filtered = df.loc[common_dates]


    df_features = df_filtered.copy()


    for other_ticker, other_df in df_dict.items():
        if other_ticker != ticker:  
            df_features[f'{other_ticker}_close'] = other_df.loc[common_dates, 'Close']

    df_features['target'] = df_features['Close'].shift(-1)  


    df_features.dropna(inplace=True)


    df_train_dict[ticker] = df_features
 
```
